```python
# importing pandas, matplotlib and seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#reading the csv file i.e our data 
data = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
data
```

```python
#Getting information about our data
data.info()
```

```python
# Number of rows and columns
data.shape
```

```python
# mapping categories of quality of sleep
sleep_quality_map = {1:'Poor',2:'Poor',3:'Poor',4:'Poor', 5:'Good',6:'Good',7:'Good',8:'Excellent',9:'Excellent', 10:'Excellent'}
data['sleep_quality'] = data['Quality of Sleep'].map(sleep_quality_map)
data
```

```python
# Viewing Quality of Sleep vs sleep_quality column which we added
data.iloc[:, 5:14:8]
```

```python
#The columns present in our data
data.columns
```

```python
# function of BMI values
def BMI_values(BMI):
    if BMI == 'Underweight':
        return 'Below 18'
    elif BMI == 'Normal' or BMI == 'Normal Weight':
        return '18 to 25'
    elif BMI == 'Overweight':
        return '25 to 30'
    elif BMI == 'Obese':
        return 'Above 30'
```

```python
#Applying our function
data['BMI_Values'] = data['BMI Category'].apply(BMI_values)
data.head(10)
```

```python
# information of the data
data.info()
```

```python
# mean of the column called Age
data['Age'].mean()
```

```python
# standard deviation of column called Age
data['Age'].std()
```

```python
# Boxplot of the column Age using matplotlib
plt.figure(figsize=(10,8))
plt.boxplot(data['Age'])
plt.title('Age Distribution')
plt.ylabel('Ages')
plt.show()
```

```python
#Histogram of column Age using matplotlib
plt.figure(figsize = (10,8))
plt.hist(data['Age'], bins=10, label = "ages", color = 'pink')
plt.title('Age Distribution')
plt.xlabel('Ages')
plt.ylabel('Frequencies')
plt.legend(loc = 2)
plt.show()
```

```python
# 2 line plots(Sleep Duration vs Quality of Sleep and Sleep Duration vs Stress Level) using matplotlib
fig, ax = plt.subplots(figsize = (18,8))
ax.plot(data['Sleep Duration'], data['Quality of Sleep'], color = 'purple', lw = 2, ls = ':')
ax.plot(data['Sleep Duration'], data['Stress Level'], color = 'green', ls = '-.')
ax.set_title('Sleep Duration vs Quality of Sleep / Stress Level')
ax.set_xlabel("Sleep Duration")
ax.set_ylabel("Quality of Sleep / Stress Level")
ax.legend(["Duration vs Quality", "Duration vs Stress"], loc = 3)
```

```python
# Bargraph of Occupation vs Sleep Duration using seaborn
sns.set(rc={'figure.figsize':(22,6)}) 
sns.set_context('talk')
sns.set_style('darkgrid')
sns.barplot(data['Occupation'], data['Sleep Duration'])
```
